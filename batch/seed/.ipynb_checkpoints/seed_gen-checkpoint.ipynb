```python
# Staff
from faker.factory import Factory
import csv
import random
import datetime

Faker = Factory.create
fake = Faker()
fake.seed(0)
fake = Faker("ja_JP")

rows = [
    ["上村 陽平", "kurupeku@example.com"],
    ["ゲスト", "guest@example.com"],
    ["佐藤 薫", "sato@example.com"],
    ["杉下 浩一", "sugishita@example.com"],
    ["山内 亮", "yamauchi@example.com"],
    ["椎名 里香", "shina@example.com"],
    ["小阪 裕太", "kosaka@example.com"],
    ["岡本 結衣", "okamoto@example.com"],
    ["安田 和哉", "yasuda@example.com"],
]

with open("csv/staffs.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)
    
    writer.writerow(["name", "email"])
    writer.writerows(rows)
    

# User & Address
amount = 1000

companies = []
for i in range(amount):
    if i % 3 == 0:
        companies.append("")
    else:
        companies.append(fake.company())
        
users = []

with open("csv/users.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)

    writer.writerow(["company_name", "name", "email", "tel"])
    for i in range(amount):
        company = random.choice(companies)
        phone = "0" + fake.phone_number()[1:]
        if company == "":
            email = fake.free_email()
        else:
            email = fake.company_email()
        row = [company, fake.name(), email, phone]
        users.append(row)
        writer.writerow(row)

with open("csv/addresses.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)

    writer.writerow(["postal_code", "prefecture", "city", "street", "building"])
    for i in range(amount):
        town = fake.town()
        street = fake.town() + fake.chome() + fake.ban()
        if i % 2 == 0:
            building_name = town + fake.building_name() if i % 4 == 0 else fake.building_name() + town
            building = building_name + fake.building_number()
        else:
            building = ""
        writer.writerow([fake.postcode(), fake.prefecture(), fake.city(), street, building])
        

# Menu
names = [
    "コンサルティング",
    "自社システム開発",
    "クラウド開発・移行サポート",
    "開発プロジェクト管理",
    "データベース設計構築",
    "システム運用・管理",
    "システムテスト・DevOps",
    "開発スタッフ派遣"
]
today = datetime.date.today()

with open("csv/menus.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)
    
    writer.writerow(["name", "published_on"])
    
    for name in names:
        writer.writerow([name, today])
        
        
# Inquiry & Progress
amount = 1100

it_list = ["導入時期未定", "3年以内", "1年以内", "なるべく早く", "開発・運用中"]
rank_list = [0, 1, 2, 3]
state_list = ["waiting", "waiting_recontact", "contacting", "estimating", "archived", "ordered"]
today = datetime.date.today()

with open("csv/inquiries.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)

    writer.writerow(["company_name", "name", "email", "tel", "number_of_users", "introductory_term", "detail"])
    for i in range(amount):
        
        if len(users) > i:
            user_attr = users[i]
        else:
            user_attr = [fake.company(), fake.name(), fake.company_email(), "0" + fake.phone_number()[1:]]
        
        nou = random.randint(0,1000) // 10 * 10
        row = user_attr + [nou, random.choice(it_list), fake.text(max_nb_chars=200)]
        writer.writerow(row)
        
with open("csv/progresses.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)

    writer.writerow(["rank", "state"])
    for i in range(amount):
        rank = random.choice(rank_list)
        state = random.choice(state_list)
        writer.writerow([rank, state])
        

# Comment
amount = 3300

with open("csv/comments.csv", "w") as f:
    writer = csv.writer(f, lineterminator="\n")
    f.truncate(0)

    writer.writerow(["content"])
    for i in range(amount):
        writer.writerow([fake.text(max_nb_chars=80)])
```
